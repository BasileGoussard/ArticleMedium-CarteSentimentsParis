```python
import glob
import os 
import json
import lzma
import numpy as np
import pandas as pd
```

```python
#Hastag
# directory = 'eolienne'
directory = 'parisstreet'


file_list = glob.glob('./'+directory+'/preprocessing'+ '/*.xz')
len(file_list)
```

```python
# All hashtag

hashtag = dict()

for file in file_list:
    file_json = json.load(lzma.open(file))
    edge_media = file_json['node']['edge_media_to_caption']
    if edge_media :
        edge_media_edge = edge_media['edges']
        if edge_media_edge:
            node = edge_media_edge[0]['node']
            if node : 
                text = node['text']
                if text:
                    phrase = ''
                    list_text = text.split()
                    for mot in list_text:
                        phrase = phrase + ' ' + mot
                    hashtag[file] = phrase
```

```python
hashtag
```

# NLP

```python
file_list = hashtag.keys()
```

```python
from afinn import Afinn
af = Afinn()

hastag = dict()
for file in file_list:
    score = af.score(hashtag[file])
    hastag[file] = score
    
df_hastag = pd.DataFrame.from_dict(hastag, orient = 'index', columns = ['sentiment'])
df_hastag = df_hastag.reset_index()
```

```python
output_path = directory

df_hastag.to_csv(os.path.join(output_path,'sentiment.csv'))
```
